```python
directors = df['director'].value_counts()
directors.head(10)
```

```python
# From the above cell we can say that for 2588 rows director names are not given so let's find out how many names were given
```

```python
given_directors = directors.sum() - directors[0]
print(f'given directors = {given_directors}')
```

```python
#Pie chart between number of director's name given and director's name not given
```

```python
# Sample data for demonstration
data = {'Directors': ['Given Directors', 'Not Given Directors'], 'Count': [directors[0], given_directors]}
df1 = pd.DataFrame(data)

# Creating the pie chart using Pandas
df1.set_index('Directors').plot.pie(y='Count', 
                                   figsize=(8, 8), 
                                   autopct='%1.1f%%', 
                                   colors=['grey', 'darkred'],
                                   title='Given Directors Vs Not Given Directors')

# Display the figure
plt.ylabel('')  # Hide the y-label
plt.show()
```

```python
df['type'].value_counts()
```

```python
# Create a pivot table for plotting
pivot_table = x.pivot(index='country', columns='type', values='count').fillna(0)

# Plotting
pivot_table.plot(kind='bar', figsize=(10, 6))

# Tilt x-axis labels by 45 degrees
plt.xticks(rotation=45)

# Set labels and title
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Count of Countries by Type')
plt.legend(title='Type')

plt.show()
```

```python
ratings = df.groupby(['rating', 'type'])['type'].value_counts().reset_index()
ratings
```

```python
# Top 10 movie ratings with maximum occurence
```

```python
top_10_TV_Show_ratings = ratings[ratings['type'] == 'TV Show'].sort_values(by='count', ascending=False).iloc[0:10]
top_10_TV_Show_ratings
```
